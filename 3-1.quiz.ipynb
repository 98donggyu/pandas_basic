## 4. 분양가격 column을 int로 변환 (실전 전처리)

### 4-1 '분양가격' 컬럼에 '   ' 공백이 있는 데이터들이 있는지 확인해 보기

```python
df[df["분양가격"].astype(str).str.strip() == ""]
```

### 4-2. '분양가격' column에 공백 제거하고 확인하기

```python
df["분양가격"] = df["분양가격"].astype(str).str.strip()
df[df["분양가격"] == ""]
```

### 4-3 '분양가격' 컬럼에 공백이 있는경우, 0으로 대입하고, int형으로 변경해보기

```python
df.loc[df["분양가격"] == "", "분양가격"] = "0"
```

```python
df["분양가격"] = df["분양가격"].astype(int)
```

### 4-4 NaN 값은 0으로 채우고, 다시 int로 형변환하기

```python
df.loc[df["분양가격"] == "nan", "분양가격"] = "0"
```

```python
df["분양가격"] = df["분양가격"].astype(int)
```

```python
df.loc[df["분양가격"] == "-"]
```

```python
df["분양가격"] = df["분양가격"].str.replace("-", "")
```

```python
df["분양가격"] = df["분양가격"].astype(int)
```

```python
df_house = pd.read_csv(file_path)
df_house
```

## 7. 지역명을 그룹으로 평균 분양가격 확인하기

## 8. 분양가격이 100보다 작은 행은 제거해 보기

### 8-1. 분양가격이 100보다 작은 값 확인하기

```python
df_house.loc[df["분양가격"] < 100]
```

```python
idx = df_house.index[df_house["분양가격"] < 100]
idx
```

### 8-2. index를 행 기준으로 drop하기

```python
df_house = df_house.drop(index=idx)
```

```python
df_house.count()
```

## 9. 다시 한 번 지역명을 group으로 묶어 확인하기

### 9-1. 지역명을 group으로 묶어 분양가격의 평균 확인하기

```python
df_house.groupby("지역명")["분양가격"].mean().sort_values(ascending=False)
```

```python
df_house.groupby("지역명")["분양가격"].count()
```

```python
df_house.groupby("지역명")["분양가격"].count().sort_values(ascending=False)
```

## 10. 연도별로 분양가격 확인하기

```python
df_house.groupby("연도")["분양가격"].mean()
```

## 11. 피벗 테이블 활용해보기

```python
df_house_pt = pd.pivot_table(df_house, index = "연도", columns = "규모구분", values = "분양가격")
df_house_pt
```

```python
df_house_pt.round(2)
```

## 12. 년도에 따른 평형대의 변화 추이를 보기위해 시각화 하기

## 13. 연도별, 규모별 가격을 알아볼까요? (multi-index)
 - 평균의 출력 형태는 소숫점 2자리까지

```python
df_house_year = df_house_pt.stack().round(2)
df_house_year
```

## 14. 위의 결과를 보기좋게 하기위해 데이터프레임으로 표현하기
 - 분양가격은 소숫점 2자리까지 표시하기

```python
df_house_year_df = pd.DataFrame(df_house_year)
df_house_year_df.columns = ["분양가격"]
df_house_year_df
```
